##Torres de Hanoi con Divide y vencerás

```
def Torres_Hanoi(N, desde, hasta):
  if N ==1 :
    print("Lleva la ficha " ,desde , " hasta ", hasta )

  else:
    #Torres_Hanoi(N-1, desde, 6-desde-hasta ) desde el principio hasta pivote
    Torres_Hanoi(N-1, desde, 6-desde-hasta )
    print("Lleva la ficha " ,desde , " hasta ", hasta )
    #Torres_Hanoi(N-1,6-desde-hasta, hasta ) desde pivote hasta final
    Torres_Hanoi(N-1, 6-desde-hasta  , hasta )


Torres_Hanoi(3, 1 , 3)

"""
Pivote --> posición 2
La función usa dos llamadas recursivas
en el primero, desde se queda igual (1) pero hasta ahora es 2
¿Por qué 6? 1 + 2 +3
Para mover los discos, el algoritmo necesita saber cuál es el poste que queda libre para usarlo como puente.
En lugar de usar muchos if/else, se usa la resta: 6 - desde - hasta
 6-2-3 = 1 (pivote es 1)
 6-1-3 = 2
 6-1-2 = 3
EJEMPLO: Torres_Hanoi(2,1,2)
          --> Torres_Hanoi(1,1,3) primera llamada recursiva
          --> Torres_Hanoi(1,3,2) 2ª llamada recursiva: Lleva la ficha 3 hasta 2

OTRO EJ: Torres_Hanoi(2,2,3)
         --> Torres_Hanoi(1,2,1)
         --> Torres_Hanoi(1,1,3)
"""
```

```
#Sucesión_de_Fibonacci
#https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci
#Calculo del termino n-simo de la suscesión de Fibonacci
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

print(Fibonacci(10))


#Complejidad de usar recursividad en Fibonacci: exponencial O(2ⁿ)
```

##PROPUESTA DE MEJORA PARA SUCESIÓN DE FIBONACCI

```
#PROPUESTA DE MEJORA PARA LA SUCESIÓN DE FIBONACCI

#El objetivo es crear una función no recursiva para pasar de la complejidad exponencial a una más simple

def fibonacci(n: int) -> int:
    if n < 2:
        return 1

    a, b = 1, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

print(fibonacci(1000))


#La función recursiva anterior es más lenta dado que internamente hace un mayor número de operaciones y además no permite números grandes,
#como por ejemplo 1000. Esta nueva función consigue optimizar esa complejidad O(n) y calcula la serie para un número grande, evitando la recursividad.
```

##Devolución de cambio por técnica voraz

```
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM): #para que tb devuelva el índice
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])

"""
En este ejemplo, nuestro sistema monetario está compuesto por monedas de 25,10,5 y . EN ORDEN!
Calula de mayor a menor las monedas que necesitamos
Primeramente hace (15-0)/25 esto da 0, no devolvemos ninguna moneda de 25
Luego (25-0)/10 y esto da 1, ahora se actualia el valor acumulado
"""
```

## N-Reinas por técnica de vueta atrás


```
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  print(SOLUCION)
  escribe(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1: #ejemplo array sol: (0,1,2,2) si >1 sol no viable pq no pueden estar en la misma fila
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False #(0,2,1,3) o (0,1,2,3), por ejemplo
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(4)

#construccion de un arbol de expansion. Condiciones para cada rama
#salto hacia atras (backtracking): volver al nodo padre para continuar con el desarrollo del arbol
#calcular la posicion de cada reina en el array y diferencia entre sus posiciones en valor absoluto
#si ese valor es == a la dif. entre altura, estan en la misma diagonal y por tanto sol. no valida
#fijarse en el arbol
#éste codigo guarda la solucion y sigue evaluando mas soluciones posibles
```

## Viaje por el rio. Programación dinámica

```
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS): #calcula el coste de cada ruta
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N] #nos dice cual es el siguiente mejor salto

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j): #marca los saltos a nodos intermedios
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)

#parecido al divide y vencerás pq problemas mas reducidos pero en vez de secuencia recursiva aprovechamos las
#soluciones intermedias para construir la sol final.
#Matriz de adyacencia: fila 0: origenes del nodo0. Los destinos (por ej. nodo 0) se representan en la columna 0 de la matriz.
#En la diagonal hay ceros pq desde nodo 0 hasta ese mismo nodo no hay coste.
#Inconveniente: esta matriz no es fiel al dibujo pq en la realidad no todos los nodos estan conectados los unos con los otros, sino que se llega a uno a traves de otros
#¿como representamos esto? poner un peso muy grande para que el algoritmo nunca tome ese camino por ser caro.
```
