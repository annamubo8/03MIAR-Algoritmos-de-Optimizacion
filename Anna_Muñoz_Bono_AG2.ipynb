##Programación Dinámica. Viaje por el rio
* **Definición**: Es posible dividir el problema en subproblemas más pequeños, guardando las soluciones para ser utilizadas más adelante.
* **Características** que permiten identificar problemas aplicables:<br>
  -Es posible almacenar soluciones de los subproblemas para ser utilizados más adelante<br>
  -Debe verificar el principio de optimalidad de Bellman: “en una secuencia optima de decisiones, toda sub-secuencia también es óptima” (*)<br>
  -La necesidad de guardar la información acerca de las soluciones parciales unido a la recursividad provoca la necesidad de preocuparnos por la complejidad espacial (cuantos recursos de espacio usaremos)<br>

###Problema<br>
En un río hay **n** embarcaderos y debemos desplazarnos río abajo desde un embarcadero a otro. Cada embarcadero tiene precios diferentes para ir de un embarcadero a otro situado más abajo. Para ir del embarcadero i al j, puede ocurrir que sea más barato hacer un trasbordo por un embarcadero intermedio k. El problema consiste en determinar la combinación más barata.

1.jpg

*Consideramos una tabla TARIFAS(i,j) para almacenar todos los precios que nos ofrecen los embarcaderos.<br>
*Si no es posible ir desde i a j daremos un valor alto para garantizar que ese trayecto no se va a elegir en la ruta óptima(modelado habitual para restricciones)

```
#Viaje por el rio - Programación dinámica
################################################################################

TARIFAS = [
[0,5,4,3,float("inf"),999,999],   #desde nodo 0
[999,0,999,2,3,999,11],  #desde nodo 1
[999,999, 0,1,999,4,10], #desde nodo 2
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

#999 se puede sustituir por float("inf") del modulo math
TARIFAS
```

```
#Calculo de la matriz de PRECIOS y RUTAS
#  PRECIOS - contiene la matriz del mejor precio para ir de un nodo a otro
#  RUTAS   - contiene los nodos intermedios para ir de un nodo a otro
################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]  #n x n
  RUTA = [ [""]*N for i in [""]*N]

  #Se recorren todos los nodos con dos bucles(origen - destino)
  #  para ir construyendo la matriz de PRECIOS
  for i in range(N-1):
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
```

```
PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])
```

```
#Calculo de la ruta usando la matriz RUTA
def calcular_ruta(RUTA, desde, hasta):
  if desde == RUTA[desde][hasta]:
  #if desde == hasta:
    #print("Ir a :" + str(desde))
    return desde
  else:
    return str(calcular_ruta(RUTA, desde, RUTA[desde][hasta])) +  ',' + str(RUTA[desde][hasta])

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

##Problema de Asignacion de tarea

```
#Asignacion de tareas - Ramificación y Poda
################################################################################
#    T A R E A
#   A
#   G
#   E
#   N
#   T
#   E

COSTES=[[11,12,18,40],
        [14,15,13,22],
        [11,17,19,23],
        [17,14,20,28]]

#objetivo: asignar tareas para min. costes
#COSTES --> columnas son las tareas y los agentes/proveedores las filas
#arbol de expansion completo: tiempo no polinomico, por eso ramificamos
#creamos el arbol, a cada variable final asignar un valor, y vamos al siguiente nivel hijo para asignar un siguiente
#backtracking: cuando violaba las condiciones del problema asumiamos que no merecia la pena seguir explorando y volviamos hacia atrás
#esto diferente pq si x ej. hemos asignado 2/4 tareas, es facil asignar una cota inferior y cota superior.
#Este calculo debe de producirse en un num. de operaciones simples lo mas bajo posible.
```

```
#Calculo del valor de una solucion parcial (solo asignamos un num. de agentes, aun no estamos en las hojas)
def valor(S,COSTES):
  VALOR = 0
  for i in range(len(S)):
    VALOR += COSTES[S[i]][i]
  return VALOR



valor((3,2,  ),COSTES)

#S: asignacion que hemos hecho bajando por la rama del arbol que estamos evaluando
#En este caso estamos evaluando la rama en que la tarea 1 es asignada al agente3 y la tarea 2 es asignada al agente2
# 17+17 = 34
```

```
#Coste inferior para soluciones parciales
#  (1,3,) Se asigna la tarea 1 al agente 0 y la tarea 3 al agente 1

#a partir de lo anterior, ¿como calculamos la cota inferior? Nos falta asignar al agente 0 y 1
#es decir, submatriz: [[18,40],
#                      [13,22]]
#cota inferior:

def CI(S,COSTES):
  VALOR = 0
  #Valores establecidos
  for i in range(len(S)):
    VALOR += COSTES[i][S[i]]

  #Estimacion
  for i in range( len(S), len(COSTES)   ):
    VALOR += min( [ COSTES[j][i] for j in range(len(S), len(COSTES))  ])
  return VALOR

def CS(S,COSTES):
  VALOR = 0
  #Valores establecidos
  for i in range(len(S)):
    VALOR += COSTES[i][S[i]]

  #Estimacion
  for i in range( len(S), len(COSTES)   ):
    VALOR += max( [ COSTES[j][i] for j in range(len(S), len(COSTES))  ])
  return VALOR

CI((0,1),COSTES)
```

```
#Genera tantos hijos como como posibilidades haya para la siguiente elemento de la tupla
#(0,) -> (0,1), (0,2), (0,3)
def crear_hijos(NODO, N):
  HIJOS = []
  for i in range(N ):
    if i not in NODO:
      HIJOS.append({'s':NODO +(i,)    }) #añadirla como posibilidad a expandir el arbol
  return HIJOS
```

```
crear_hijos((0,) , 4)
```

```
#calcula la cota inferior y superior y descartará aquellas que "superen" esa cota (poda)

def ramificacion_y_poda(COSTES):
#Construccion iterativa de soluciones(arbol). En cada etapa asignamos un agente(ramas).
#Nodos del grafo  { s:(1,2),CI:3,CS:5  }
  #print(COSTES)
  DIMENSION = len(COSTES)
  MEJOR_SOLUCION=tuple( i for i in range(len(COSTES)) ) #inicializar en mejor solucion que será nuestra cota a batir
  CotaSup = valor(MEJOR_SOLUCION,COSTES)
  #print("Cota Superior:", CotaSup)

  NODOS=[] #lista con nodos que nos interesa analizar
  NODOS.append({'s':(), 'ci':CI((),COSTES)    } ) #en cada nodo añadimos su asignacion y si cota inferior

  iteracion = 0

  while( len(NODOS) > 0): #mientras haya nodos que tienen el potencial de darnos una sol. mejor
    iteracion +=1

    nodo_prometedor = [ min(NODOS, key=lambda x:x['ci']) ][0]['s']
    #print("Nodo prometedor:", nodo_prometedor)

    #Ramificacion
    #Se generan los hijos
    HIJOS =[ {'s':x['s'], 'ci':CI(x['s'], COSTES)   } for x in crear_hijos(nodo_prometedor, DIMENSION) ] #añade una nueva asignacion en la siguiente posicion libre de la tupla


    #Revisamos la cota superior y nos quedamos con la mejor solucion si llegamos a una solucion final
    NODO_FINAL = [x for x in HIJOS if len(x['s']) == DIMENSION  ]
    if len(NODO_FINAL ) >0:
      #print("\n********Soluciones:",  [x for x in HIJOS if len(x['s']) == DIMENSION  ] )
      if NODO_FINAL[0]['ci'] < CotaSup:
        CotaSup = NODO_FINAL[0]['ci']
        MEJOR_SOLUCION = NODO_FINAL

    #Poda #para descartar subarboles cuyo valor es peor que el de la cota
    HIJOS = [x for x in HIJOS if x['ci'] < CotaSup   ] #coge aquellos con potencial margen de mejora

    #Añadimos los hijos
    NODOS.extend(HIJOS)

    #Eliminamos el nodo ramificado
    NODOS =  [  x for x in NODOS if x['s'] != nodo_prometedor    ]

  print("La solucion final es:" ,MEJOR_SOLUCION , " en " , iteracion , " iteraciones" , " para dimension: " ,DIMENSION  )



ramificacion_y_poda(COSTES)
```

```
import math                      #Funciones matematicas
import matplotlib.pyplot as plt  #Generacion de gráficos (otra opcion seaborn)
import numpy as np               #Tratamiento matriz N-dimensionales y otras (fundamental!)
#import scipy as sc

import random
```

Vamos a buscar el minimo de la funcion paraboloide : $$f(x) =  x² + y²$$

Obviamente se encuentra en (x,y)=(0,0) pero probaremos como llegamos a él a través del descenso del gradiante.

```
#Definimos la funcion
#Paraboloide
f  = lambda X:      X[0]**2 + X[1]**2    #Funcion
df = lambda X: [2*X[0] , 2*X[1]]         #Gradiente

df([1,2])
```

```
from sympy import symbols
from sympy.plotting import plot
from sympy.plotting import plot3d
x,y = symbols('x y')
plot3d(x**2 + y**2,
       (x,-5,5),(y,-5,5),
       title='x**2 + y**2',
       size=(10,10))
```

```
#Prepara los datos para dibujar mapa de niveles de Z
resolucion = 100
rango=5.5

X=np.linspace(-rango,rango,resolucion)
Y=np.linspace(-rango,rango,resolucion)
Z=np.zeros((resolucion,resolucion))
for ix,x in enumerate(X):
  for iy,y in enumerate(Y):
    Z[iy,ix] = f([x,y])

#Pinta el mapa de niveles de Z
plt.contourf(X,Y,Z,resolucion)
plt.colorbar()

#Generamos un punto aleatorio inicial y pintamos de blanco
P=[random.uniform(-5,5  ),random.uniform(-5,5 ) ]
plt.plot(P[0],P[1],"o",c="white")

#Tasa de aprendizaje. Fija. Sería más efectivo reducirlo a medida que nos acercamos.
TA=.1

#Iteraciones:50
for _ in range(50):
  grad = df(P)
  #print(P,grad)
  P[0],P[1] = P[0] - TA*grad[0] , P[1] - TA*grad[1]
  plt.plot(P[0],P[1],"o",c="red")

#Dibujamos el punto final y pintamos de verde
plt.plot(P[0],P[1],"o",c="green")
plt.show()
print("Solucion:" , P , f(P))
```

**¿Te atreves a optimizar la función?:**

$$f(x)=sin(1/2 * x^2 - 1/4 * y^2 + 3) * cos(2*x + 1 - e^y)$$
<br>

```
#Definimos la funcion
f= lambda X: math.sin(1/2 * X[0]**2 - 1/4 * X[1]**2 + 3) *math.cos(2*X[0] + 1 - math.exp(X[1]) )
```

```
def gradiente_numerico(X, h=1e-5):
    x, y = X[0], X[1]

    # Derivada respecto a x
    f_x_plus = f([x + h, y])
    f_x_minus = f([x - h, y])
    df_dx = (f_x_plus - f_x_minus) / (2*h)

    # Derivada respecto a y
    f_y_plus = f([x, y + h])
    f_y_minus = f([x, y - h])
    df_dy = (f_y_plus - f_y_minus) / (2*h)

    return [df_dx, df_dy]
```

```
#mapa
from sympy import symbols
from sympy.plotting import plot3d
import sympy as sp

x_sym, y_sym = symbols('x y')
funcion_sym = sp.sin(1/2 * x_sym**2 - 1/4 * y_sym**2 + 3) * sp.cos(2*x_sym + 1 - sp.exp(y_sym))
plot3d(funcion_sym, (x_sym,-3,3), (y_sym,-3,3), size=(10,10))
```

```
resolucion = 200
rango = 3
X_vals = np.linspace(-rango, rango, resolucion)
Y_vals = np.linspace(-rango, rango, resolucion)
Z = np.zeros((resolucion, resolucion))

# Calculamos los valores de f(x,y) para la malla
for ix, x in enumerate(X_vals):
    for iy, y in enumerate(Y_vals):
        Z[iy, ix] = f([x, y])

#Mapa
plt.figure(figsize=(12, 10))
contour = plt.contourf(X_vals, Y_vals, Z, resolucion, cmap='viridis', alpha=0.8)
plt.colorbar(contour, label='f(x,y)')
plt.contour(X_vals, Y_vals, Z, 20, colors='black', alpha=0.3, linewidths=0.5)

#PUNTO INICIAL ALEATORIO
P = [random.uniform(-rango+0.5, rango-0.5), random.uniform(-rango+0.5, rango-0.5)]
plt.plot(P[0], P[1], "o", c="white", markersize=10, label='Punto inicial')

#PARÁMETROS DEL DESCENSO DE GRADIENTE
TA = 0.1  # Tasa de aprendizaje
iteraciones = 100
trayectoria = [P.copy()]  # Para guardar la trayectoria

#ALG. DESCENSO DE GRADIENTE
for i in range(iteraciones):
    grad = gradiente_numerico(P)

    # Actualización de posición
    P[0] = P[0] - TA * grad[0]
    P[1] = P[1] - TA * grad[1]

    # Guardar la trayectoria
    trayectoria.append(P.copy())

    # Dibujar puntos intermedios (cada 5 iteraciones para no saturar)
    if i % 5 == 0:
        plt.plot(P[0], P[1], "o", c="red", markersize=4, alpha=0.6)

#DIBUJAR LA TRAYECTORIA COMPLETA
trayectoria = np.array(trayectoria)
plt.plot(trayectoria[:, 0], trayectoria[:, 1], 'b-', linewidth=1, alpha=0.7, label='Trayectoria')

#Punto final
plt.plot(P[0], P[1], "o", c="green", markersize=12, label='Punto final')

#CONFIGURACIÓN FINAL DEL GRÁFICO
plt.title('Descenso de Gradiente para función compleja', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.xlim(-rango, rango)
plt.ylim(-rango, rango)
plt.show()
```
